In this exercise you get a dataset of the human development index (HDI) for different countries around the world over the years. The HDI is a measure of 'how developed' a country is. Here a short description of each column in the dataset

- *Country, Year*, Region are self-explanatory
- *GDP per capita*: gross domestic product of a country per inhabitant. It is considered a measure of how rich a country is.
- *Expenditure on health/education/social protection*: percentage of public spending destined to those sectors
- *HDI*: human development index
- *Code*: three letter country code
- *Global*: Whether the country belongs to the so called 'Global North' or 'Global South'

### B: Is there a relationship between GDP per capita and HDI? is this relationship different for different regions of the world? 


```python
figure = plt.figure()
ax = figure.subplots()

region_to_index = {
    region: index for index, region in enumerate(sorted(df["Region"].unique()))
}
df["region_index"] = df["Region"].map(region_to_index)

regions_already_plotted = set()

cmap = plt.get_cmap("tab10")

for country in df["Country"].unique():
    df_country = df[df["Country"] == country]
    region = df_country["Region"].iloc[0]
    color = cmap(df_country["region_index"].iloc[0])
    
    label = None if region in regions_already_plotted else region
    regions_already_plotted.add(region)

    ax.plot(
        df_country["GDP per capita"],
        df_country["HDI"],
        linewidth=1,
        c=color,
        marker=".",
        markersize=3,
        label=label
    )
    ax.set_xlabel("GDP per capita")
    ax.set_ylabel("HDI")
    
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    
    ax.legend()

plt.show()
```

```python
figure = plt.figure()
axes = figure.subplots(2, 3)

region_to_index = {
    region: index for index, region in enumerate(sorted(df["Region"].unique()))
}
df["region_index"] = df["Region"].map(region_to_index)

regions_already_plotted = set()

cmap = plt.get_cmap("tab10")

for region_index, region in enumerate(df["Region"].unique()):
    ax = axes.flatten()[region_index]
    
    for country in df["Country"].unique():
        df_country = df[df["Country"] == country]
        country_region = df_country["Region"].iloc[0]
        
        if country_region == region:
            color = cmap(df_country["region_index"].iloc[0])
            zorder = 0
        else:
            color = "silver"
            zorder = 10
            
        label = None if country_region in regions_already_plotted else country_region
        regions_already_plotted.add(country_region)

        ax.plot(
            df_country["GDP per capita"],
            df_country["HDI"],
            linewidth=1,
            c=color,
            marker=".",
            markersize=3,
            label=label,
            zorder=zorder,
        )
        ax.set_xlabel("GDP per capita")
        ax.set_ylabel("HDI")

        ax.spines["top"].set_visible(False)
        ax.spines["right"].set_visible(False)

        if region_index == 0:
            ax.legend()

plt.show()
```
